```python
import cv2
import imutils
import dlib
import numpy as np
from imutils import face_utils
from scipy.spatial import distance
from pygame import mixer  # Import mixer properly

# Initialize pygame mixer for playing alarm sound
mixer.init()
mixer.music.load("C:\\Users\\HP\\my python stufff\\Drowsiness detection\\music.wav")

# Define EAR (Eye Aspect Ratio) function
def eye_aspect_ratio(eye):
    A = distance.euclidean(eye[1], eye[5])
    B = distance.euclidean(eye[2], eye[4])
    C = distance.euclidean(eye[0], eye[3])
    ear = (A + B) / (2.0 * C)
    return ear

# Define threshold and frame count for drowsiness detection
thresh = 0.25
flag = 0
frame_check = 15  # Number of consecutive frames to trigger alarm

# Load face detector and shape predictor
detect = dlib.get_frontal_face_detector()
predict = dlib.shape_predictor("C:\\Users\\HP\\my python stufff\\Drowsiness detection\\models\\shape_predictor_68_face_landmarks (1).dat")

# Get indexes for left and right eye landmarks
(lStart, lEnd) = face_utils.FACIAL_LANDMARKS_68_IDXS["left_eye"]
(rStart, rEnd) = face_utils.FACIAL_LANDMARKS_68_IDXS["right_eye"]

# Open webcam
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break  # Exit loop if frame is not captured

    frame = imutils.resize(frame, width=450)
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    subjects = detect(gray, 0)

    for subject in subjects:
        shape = predict(gray, subject)
        shape = face_utils.shape_to_np(shape)

        # Get eye landmarks
        leftEye = shape[lStart:lEnd]
        rightEye = shape[rStart:rEnd]

        # Compute EAR for both eyes
        leftEAR = eye_aspect_ratio(leftEye)
        rightEAR = eye_aspect_ratio(rightEye)
        ear = (leftEAR + rightEAR) / 2.0  # Average EAR

        # Draw eye contours
        leftEyeHull = cv2.convexHull(leftEye)
        rightEyeHull = cv2.convexHull(rightEye)
        cv2.drawContours(frame, [leftEyeHull], -1, (0, 255, 0), 1)
        cv2.drawContours(frame, [rightEyeHull], -1, (0, 255, 0), 1)

        # Check if eyes are closed
        if ear < thresh:
            flag += 1
            print(f"Flag: {flag}")
            if flag >= frame_check:
                cv2.putText(frame, "****************ALERT!****************", (10, 30),
                            cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)
                cv2.putText(frame, "****************ALERT!****************", (10, 325),
                            cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)
                mixer.music.play()  # Play alarm sound
        else:
            flag = 0  # Reset counter if eyes open

    # Display the frame
    cv2.imshow("Drowsiness Detector", frame)

    # Exit on pressing 'q'
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

# Cleanup
cap.release()
cv2.destroyAllWindows()
```
